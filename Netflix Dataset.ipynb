I used isnull() to identify missing values in each column. Missing data reduces data quality and must be handled before analysis.

```python
df.isnull().sum()
```

I found missing values in columns like director, cast, country and rating. Instead of deleting rows, I replaced missing values using fillna().

```python
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Not Available")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Not Rated")
df["duration"] = df["duration"].fillna("Unknown")
```

The date column also had missing values, so I filled those missing entries first before converting the column into date format.”

```python
df["date_added"] = df["date_added"].fillna("January 1, 2000")#Handled missing values using fillna() to avoid data loss.
```

Verified that missing values were handled successfully using isnull().

```python
df.isnull().sum()
```

After handling missing values, I checked duplicate rows using duplicated().

I removed duplicate rows using drop_duplicates() and checked the dataset shape again.

```python
duplicates = df.duplicated().sum()#remove repeated records
print("Duplicate Rows:", duplicates)
```

Removed duplicate rows and verified dataset size again.

```python
df = df.drop_duplicates()
print("After Removing Duplicates:", df.shape)
```

Standardized text values to maintain consistency in dataset.

```python
df["type"] = df["type"].str.lower()
df["country"] = df["country"].str.title()
```

Converted date format into consistent dd-mm-yyyy format.

```python
# Convert date safely
df["date_added"] = pd.to_datetime(
    df["date_added"],
    errors="coerce"
)

# Fill dates that became invalid after conversion
df["date_added"] = df["date_added"].fillna(pd.Timestamp("2000-01-01"))

# Format date consistently
df["date_added"] = df["date_added"].dt.strftime("%d-%m-%Y")
```

```python
print(df.isnull().sum())
```

Renamed column headers by converting into lowercase and removing spaces.(I cleaned column names by converting them into lowercase and replacing spaces with underscores.)

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

Performed final missing value check after cleaning process.

```python
print("Final Missing Values:")
print(df.isnull().sum())
```
